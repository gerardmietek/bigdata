```
from pyspark.sql.functions import col
filePath = "dbfs:/FileStore/tables/Files/names.csv"
namesDf = spark.read.format("csv").option("header","true").option("inferSchema","true").load(filePath)
display(namesDf.limit(10))
```

```
namesDf = namesDf.fillna({"death_details": "UNKNOWN"})
display(namesDf.limit(10))
#działa
```

```
print("before dropping nulls: ", namesDf.count())
namesDf = namesDf.dropna()
print("after dropping nulls: ", namesDf.count())
```

```
from pyspark.sql.functions import expr

namesDf = namesDf.withColumn("name", expr("NULLIF(name, 'Fred Astaire')"))
display(namesDf.limit(10))
```

```
namesDf = namesDf.withColumn("name", expr("IFNULL(name, 'NONAME')"))
display(namesDf.limit(10))
```

```
from pyspark.sql.functions import regexp_replace
namesDf = namesDf.withColumn("name", regexp_replace(col("name"), "a", "X"))
display(namesDf.limit(10))
```

```
namesDf = namesDf.replace({177: 666}, subset=["height"])
display(namesDf.limit(10))
```

```
from pyspark.sql.functions import explode, split
namesDf = namesDf.withColumn("birth_name", split(col("birth_name"), " "))
```

```
from pyspark.sql.functions import array_contains
namesDf = namesDf.withColumn("name_contains_fred", array_contains(col("birth_name"), "Fred"))
display(namesDf.limit(10))
```

```
namesDf = namesDf.withColumn("birth_name", explode(col("birth_name")).alias("name_parts"))
display(namesDf.limit(10))
```

```
from pyspark.sql.functions import regexp_extract
namesDf = namesDf.withColumn("first_h_number", regexp_extract(col("height"), r"(\d)", 1))
display(namesDf.limit(10))
```

```
#Agregujące
from pyspark.sql.types import IntegerType, FloatType
df_int = namesDf.filter(col("height").cast(IntegerType()).isNotNull())
suma = df_int.agg(sum("height"))

#szczerze nie mam pojecia co zrobic
```

```
from pyspark.sql.functions import udf
from pyspark.sql.types import IntegerType, StringType
import pandas as pd
from pyspark.sql.functions import pandas_udf

def divorceX2(divorce):
    if divorce is None:
        return None
    return divorce * 2

divorce_udf = udf(divorceX2, IntegerType())
namesDf = namesDf.withColumn("divorceX2", divorce_udf( col("divorces")))
display(namesDf.limit(10))
```

```
@pandas_udf(StringType())
def to_lower_case(name_series: pd.Series) -> pd.Series:
    return name_series.str.lower()

namesDf = namesDf.withColumn("birth_name", to_lower_case(col("birth_name")))
display(namesDf.limit(10))
```
